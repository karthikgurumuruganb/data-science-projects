```python
#new_columns=['age', 'cp', 'trestbps', 'chol', 'thalach', 'exang', 'oldpeak', 'ca'] # Chi2
new_columns=['sex', 'cp', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal'] # correlation
x = df[new_columns]
x.head()
```

```python
x = pd.get_dummies(x, columns=['sex','cp','exang', 'slope','ca', 'thal'])
x.head()
```

```python
#SCALING to be done for thalach and oldpeak
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['thalach', 'oldpeak']
x[columns_to_scale] = standardScaler.fit_transform(x[columns_to_scale])

x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
y_test.size
```

Using KNN Classifier

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model = KNeighborsClassifier()
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

Decision Tree Classifier

```python
model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
model1.score(x_train, y_train)
```

Random Forest Classifier

```python
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2.score(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted, normalize=True)
```

```python
y_predicted = model1.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted, normalize=True)
```

```python
y_predicted = model2.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted, normalize=True)
```
